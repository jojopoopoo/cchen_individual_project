The Platform in df contains both handheld and console video game data. However, we do not want to mix the two types of devices because in my opinion, it is not fair to mix these two types of devices becuase they are very different in terms of usage, size, price, etc. Therefore, we need to recognize the data associated with console video game platform and drop the ones for handheld.

## Video Game Platform (Console, not handheld) unit sales data

```python
df2 = pd.read_csv('platform_sales.csv',index_col=0)
```

```python
df2.head()
```

```python
import re
```

```python
df2['Platform'] = df2['Platform'].str.extract(r"\((.*?)\)", expand=False)
```

```python
df2.head()
```

```python
df2.to_csv('new_platform_final.csv', index=False)
```

Drop the video game sales data of portable devices, keep only console game data.

```python
df = df[df.Platform.isin(df2.Platform)]
```

```python
df.head(10)
```

Reindex the dataframe to start from 1

```python
df.index = np.arange(1, len(df) + 1)
```

```python
df.head(10)
```

```python
df.dropna(inplace=True)
```

```python
df['Year'] = df['Year'].astype(int)
```

```python
df['Year'] = df['Year'].astype(str)
```

```python
df.head()
```

```python
df.to_csv('new_data_final.csv', index=False)
```
